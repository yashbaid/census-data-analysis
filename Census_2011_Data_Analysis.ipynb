Import Python Libraries

```python
import warnings
warnings.filterwarnings('ignore')

import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',500)
plt.rcParams["figure.figsize"] = [20,10]
```

Import dataset for Census 2011 

```python
Census_Data = pd.read_csv('C:\Kaggle\Census 2011\\india-districts-census-2011.csv', encoding='iso-8859-1')
```

Changing name of few columns.

```python
Census_Data = Census_Data.rename(columns = {"State name": "State", "District code":"District_Code", "District name": "District"}) 
```

Below piece of code will calculate percentage of missing values in each column.

```python
print(100*(Census_Data.isnull().sum()/len(Census_Data.index)))
```

Below piece of code will group the data statewise.

```python
Census_Data=Census_Data.groupby(['State']).sum().sort_values(by='State').reset_index()
```

```python
Census_Data.head(10)
```

Let's sort states by highest to lowest population.

```python
Census_Data.sort_values(by='Population', ascending=False)
```

Let's understand the population distribution between states more clearly with the help of below graph.

```python
Census_Data.sort_values(by='Population', ascending=True).plot.barh(x='State', y='Population', title='Population distribution between states in India')
plt.xscale('log')
plt.xlabel('Population')
plt.ylabel('States')
plt.show()
```

Let's understand the population distribution between gender in whole country with the help of below graph.

```python
Census_Data['Country']='India'

Census_Data['Percent_Female']=round((Census_Data['Female']/Census_Data['Population'])*100,2)
Census_Data['Percent_Male']=round((Census_Data['Male']/Census_Data['Population'])*100,2)

Census_Data.groupby(['Country'])['Percent_Female','Percent_Male'].mean().reset_index().plot(kind ='barh', title='Population distribution between gender in India')




```

Below piece of code gives us an idea there is one outlier state in terms of population, this state is Uttar Pradesh as the describe() function used below shows us Uttar Pradesh is the state with maximum population.

```python
plt.boxplot(Census_Data['Population'])
plt.yscale('log')
plt.ylabel('Population')
plt.title('Outlier state in terms of population')
plt.show()
Census_Data['Female'].describe()
```

Let's understand which state has the maximum literacy rate through below piece of code.


```python
Census_Data['Literacy_Rate']=round((Census_Data['Literate']/Census_Data['Population'])*100,2)
Census_Data['Literacy_Rate_Male']=round((Census_Data['Male_Literate']/Census_Data['Male'])*100,2)
Census_Data['Literacy_Rate_Female']=round((Census_Data['Female_Literate']/Census_Data['Male'])*100,2)

Census_Data.reset_index().sort_values(by='Literacy_Rate',ascending=False)
```

Output for above piece of code tells us Kerala has highest literacy rate (84%).

Now's let's understand comparison of literacy rate between states using below graph.

```python
Census_Data.sort_values(by='Literacy_Rate', ascending=True).plot.barh(x='State', y=['Literacy_Rate'], title='Literacy Rate comparison between states in India')
plt.xscale('log')
plt.xlabel('Population')
plt.ylabel('States')
plt.show()
```

Now let's compare the literacy rate of male and female in every state using below graph.

```python
Census_Data.sort_values(by='Literacy_Rate', ascending=True).plot.barh(x='State', y=['Literacy_Rate_Male','Literacy_Rate_Female'], title='Literacy rate of male and female in every state')
plt.xscale('log')

plt.xlabel('Literacy Rates')
plt.ylabel('States')
plt.figure(figsize=(35,25))

plt.show()
```

Below graph shows us literacy rate of males is much more as compared to that of females in India.

```python
print(Census_Data.groupby(['Country'])['Literacy_Rate_Female','Literacy_Rate_Male'].sum().reset_index().plot(kind ='barh', title='Literacy Rate distribution between gender in India'))
```

```python
Census_Data['Households_Without_LPG_OR_PNG']=Census_Data['Households']-Census_Data['LPG_or_PNG_Households']

Census_Data.sort_values(by='LPG_or_PNG_Households',ascending=True).plot.barh(x='State',y=['Households_Without_LPG_OR_PNG','LPG_or_PNG_Households'])
```

Let's see distribution of households having mobile phone and households having internet facility.

```python
Census_Data.sort_values(by='Households_with_Internet',ascending=True).plot.barh(x='State',y=['Households_with_Telephone_Mobile_Phone_Mobile_only','Households_with_Internet'],title='Distribution of households having mobile phone and households having internet facility')
```

Above graph is so surprising as in every state most of the households despite of having mobile phones still do not have internet facility.

Let's see distribution of households with mobile phones and without mobile phones in each and every state.

```python
Census_Data['Households_With_Mobile']=Census_Data['Households_with_Telephone_Mobile_Phone_Mobile_only']+ Census_Data['Households_with_Telephone_Mobile_Phone_Both']
Census_Data['Households_Without_Mobile']=Census_Data['Households']-Census_Data['Households_With_Mobile']

Census_Data.sort_values(by='Households_With_Mobile',ascending=True).plot.barh(x='State',y=['Households_Without_Mobile','Households_With_Mobile'],title='Distribution of households with mobile phones and without mobile phones in each and every state')
plt.xlabel('Households')
plt.ylabel('States')
```

The above graph shows almost more than half of households in evey state have mobile phones. Well Kerala, TamilNadu and Punjab are doing great.
